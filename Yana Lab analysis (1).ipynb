```python
def ValueIsGood(val):
    if pandas.isnull(val):
        return False
    
    val = val.lower()
    
    if val.find("human") != -1:
        return True
    
    return False
    


covdata_df["GoodOrigin"] = [ValueIsGood(covdata_df["Origin"][i]) for i in range(len(covdata_df))] #create new column
print(covdata_df["GoodOrigin"])
```
