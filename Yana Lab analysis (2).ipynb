```python
#maybe-RBD
def ValueIsGood(val):
    if pandas.isnull(val):
        return False
    
    if val=="S; RBD/non-RBD" or val=="S; S1/S2"or val=="S; probably RBD (implied by clustering)" or val=="S; Possibly RBD" or val=="S; S1/S2 Cleavage Site" or val=="S; S1":
        return True
    
    return False
    


covdata_df["Maybe-RBD"] = [ValueIsGood(covdata_df["Protein + Epitope"][i]) for i in range(len(covdata_df))] #create new column
print(covdata_df["Maybe-RBD"])
```
